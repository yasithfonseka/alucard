**COVID-19 VISUALIZATION**

```
import pandas as pd
from matplotlib import pyplot as plt

plt.rcdefaults()

data = pd.read_csv('research_time_series_covid19_recovered_sothasia.csv')

daycount = data['Day']

AF = data['Afghanistan']
IN = data['India']
PAK = data['Pakistan']
BAN = data['Bangladesh']
SL = data['Sri Lanka']
NP = data['Nepal']
BU = data['Bhutan']
MD = data['Maldives']

colors = ['#1f77b4', '#aec7e8', '#ff7f0e', '#ffbb78', '#2ca02c', '#800080', '#FFFF00', '#CD5C5C']

plt.plot(daycount, AF, color=colors[0], label="Afghanistan")
plt.plot(daycount, IN, color=colors[1], label="India")
plt.plot(daycount, PAK, color=colors[2], label="Pakistan")
plt.plot(daycount, BAN, color=colors[3], label="Bangladesh")
plt.plot(daycount, SL, color=colors[4], label="Sri Lanka")
plt.plot(daycount, NP, color=colors[5], label="Nepal")
plt.plot(daycount, BU, color=colors[6], label="Bhutan")
plt.plot(daycount, MD, color=colors[7], label="Maldives")

plt.legend()
plt.title('COVID-19 NUMBER OF RECOVERED IN SOUTH ASIA WITHIN 114 DAYS')
plt.ylabel('Number of recovered')
plt.xlabel('Number of days')

plt.savefig("covid19_southasia_recovered.png")

plt.show()
```
